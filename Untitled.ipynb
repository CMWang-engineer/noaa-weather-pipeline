```python
import pandas as pd

df = pd.read_csv('isd-history.csv')
print(df.shape)
print(df.columns.tolist())
print(df['CTRY'].value_counts().head(20))
```

```python
# 筛选主要国家 + 2020年后还活跃的站点
major_countries = ['US', 'CA', 'UK', 'FR', 'GM', 'JA', 'IN', 'BR', 'AU', 'CH']

df['END'] = pd.to_datetime(df['END'], format='%Y%m%d')

active_stations = df[
    (df['CTRY'].isin(major_countries)) &
    (df['END'] >= '2023-01-01') &
    (df['LAT'] != '+00.000')  # 去掉经纬度为空的
]

print(active_stations.shape)
print(active_stations['CTRY'].value_counts())
```

```python
selected_stations = active_stations.groupby('CTRY').head(20).copy()

selected_stations['STATION_ID'] = (
    selected_stations['USAF'].astype(str).str.strip('"') +
    selected_stations['WBAN'].astype(str).str.zfill(5)
)

print(selected_stations.shape)
print(selected_stations[['STATION_ID', 'STATION NAME', 'CTRY']].head(10))

selected_stations[['STATION_ID', 'STATION NAME', 'CTRY', 'LAT', 'LON']].to_csv('selected_stations.csv', index=False)
print("已保存 selected_stations.csv")
```

```python
import os
import time

years = [2020, 2021, 2022, 2023]
base_url = "https://www.ncei.noaa.gov/data/global-summary-of-the-day/access"
os.makedirs('raw_data', exist_ok=True)

station_ids = selected_stations['STATION_ID'].tolist()

total = len(station_ids) * len(years)
count = 0
failed = []

for year in years:
    os.makedirs(f'raw_data/{year}', exist_ok=True)
    for station_id in station_ids:
        url = f"{base_url}/{year}/{station_id}.csv"
        output = f"raw_data/{year}/{station_id}.csv"
        
        if os.path.exists(output):
            count += 1
            continue
            
        result = os.system(f'curl -s -f -o {output} {url}')
        
        if result != 0:
            failed.append(f"{year}/{station_id}")
            os.remove(output) if os.path.exists(output) else None
        
        count += 1
        if count % 50 == 0:
            print(f"进度: {count}/{total}")
        
        time.sleep(0.1)  # 避免请求太快

print(f"\n完成！失败: {len(failed)} 个")
print(failed[:10] if failed else "全部成功")
```

```python
import pandas as pd
import os

dfs = []

for year in [2020, 2021, 2022, 2023]:
    for file in os.listdir(f'raw_data/{year}'):
        df = pd.read_csv(f'raw_data/{year}/{file}')
        dfs.append(df)

df_all = pd.concat(dfs, ignore_index=True)

print(f"总行数: {df_all.shape[0]:,}")
print(f"总列数: {df_all.shape[1]}")
print(f"\n缺失值统计:")
print(df_all.isin([999.9, 9999.9, 99.99]).sum())
```

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = SparkSession.builder \
    .master("local[*]") \
    .appName("noaa_processing") \
    .getOrCreate()

df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("noaa_raw_2020_2023.csv")

df.printSchema()
```

```python
# 1. 只保留有用的列
cols_to_keep = ['STATION', 'DATE', 'LATITUDE', 'LONGITUDE', 
                'ELEVATION', 'NAME', 'TEMP', 'MAX', 'MIN', 
                'PRCP', 'WDSP']

df_clean = df.select(cols_to_keep)

# 2. 替换缺失值（999.9 → null）
df_clean = df_clean \
    .withColumn('TEMP', F.when(F.col('TEMP') == 999.9, None).otherwise(F.col('TEMP'))) \
    .withColumn('MAX',  F.when(F.col('MAX')  == 999.9, None).otherwise(F.col('MAX'))) \
    .withColumn('MIN',  F.when(F.col('MIN')  == 999.9, None).otherwise(F.col('MIN'))) \
    .withColumn('PRCP', F.when(F.col('PRCP') == 99.99, None).otherwise(F.col('PRCP'))) \
    .withColumn('WDSP', F.when(F.col('WDSP') == 999.9, None).otherwise(F.col('WDSP')))

# 3. 华氏度转摄氏度
df_clean = df_clean \
    .withColumn('TEMP_C', F.round((F.col('TEMP') - 32) * 5 / 9, 2)) \
    .withColumn('MAX_C',  F.round((F.col('MAX')  - 32) * 5 / 9, 2)) \
    .withColumn('MIN_C',  F.round((F.col('MIN')  - 32) * 5 / 9, 2))

# 4. 提取国家代码
df_clean = df_clean.withColumn('COUNTRY', F.split(F.col('NAME'), ', ').getItem(1))

# 5. 添加年月字段
df_clean = df_clean \
    .withColumn('YEAR',  F.year(F.col('DATE'))) \
    .withColumn('MONTH', F.month(F.col('DATE')))

df_clean.show(3)
print(f"总行数: {df_clean.count():,}")
```

```python
# 保存为 parquet（按年分区）
df_clean.write \
    .mode("overwrite") \
    .partitionBy("YEAR") \
    .parquet("noaa_processed")

print("保存完成！")
```
